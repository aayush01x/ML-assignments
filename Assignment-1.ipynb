```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import product
```

```python
def findpairs(points, r):
    new_dim = {}
    for i, pt in enumerate(points):
        change = tuple(int(k// r) for k in pt)
        if change in new_dim: new_dim[change].append(i)
        else:new_dim[change] = [i]
    pair= []
    for i, pt in enumerate(points):
        change = tuple(int(k // r) for k in pt)
        for diff in product([-1, 0, 1], repeat=len(pt)):
            adj = tuple(x+y for x,y in zip(change, diff))
            if adj in new_dim:
                for j in new_dim[adj]:
                    if np.linalg.norm(points[i] - points[j]) <= r and i!=j:pair.append((points[max(i, j)], points[min(i, j)]))
    return pair

def solver(d):
    np.random.seed(101)
    points = np.random.uniform(0,100,size=(1000, d))
    A= []
    if(d==2):
        plt.scatter(points[:, 0], points[:, 1])
    for r in range(1,10):
        A.append(findpairs(points,r))
    return A
```

# d=2

```python
solver(2)
```
